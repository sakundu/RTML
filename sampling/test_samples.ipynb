# Sampling Architectural Configurations for Axiline designs

## Sampling for Axiline-SVM, Logistic, Linear

### Sample space details

```python
space = [(2, 25), (5, 60)]
x_range = [1, 26]
y_range = [4, 62]
x_mark = [2, 1]
y_mark = [5, 1]
x_label = 'Number of Cycle'
y_label = "Size"
test = [[4, 50], [5, 40], [8, 25], [10, 20], [11, 35], [15, 25], [20, 10], \
        [20, 45], [5, 20], [10, 50]]

valid = [[15, 40], [20, 36], [10, 45], [18, 25], [17, 11], [7, 20], [13, 28], \
        [5, 49], [12, 15], [8, 32]]
```

### Sobol sampling

```python
## Sobol Sampling
sobol_samples_16 = gen_sobol_sample(16, space, 2)
sobol_samples_32 = gen_sobol_sample(32, space, 2)
print(sobol_samples_16)
print(sobol_samples_32)
ax = plot_sample(sobol_samples_32, 'Train', 'ob', None, x_range, y_range, 
                 x_mark, y_mark, x_label, y_label, 
                 'Sampling Using Sobol Sequence. Sample size: 32')

ax = plot_sample(test, 'Test', 'or', ax)
ax = plot_sample(valid, 'Validation', 'oc', ax)
# ax = plot_sample(valid, 'Valid', 'oc', ax)
_ = ax.grid()
_ = ax.legend()
plt.show()
check_overlap(sobol_samples_32, test)
check_overlap(sobol_samples_32, valid)
```

### Halton sampling

```python
## Halton Sampling, seed does not affect the outcome
seed = 5
halton_samples_16 = gen_halton_sample(16, space, seed)
halton_samples_32 = gen_halton_sample(32, space, seed)
print(halton_samples_16)
print(halton_samples_32)
ax = plot_sample(halton_samples_32, 'Training', 'ob', None, x_range, y_range, 
                x_mark, y_mark, x_label, y_label, 
                'Sampling Using Halton Sequence. Sample size: 32')

ax = plot_sample(test, 'Test', 'or', ax)
ax = plot_sample(valid, 'Validation', 'oc', ax)
_ = ax.grid()
_ = ax.legend(fontsize='small')
plt.show()
check_overlap(halton_samples_32, test)
check_overlap(halton_samples_32, valid)
```

### LHS sampling for validation data set generation

```python
# Our plan to use these LHS sample data points as validation set during training
validation_space = [(4, 21), (9, 51)]
lhs_samples_10 = gen_lhs_sample(10, validation_space, 0, 1, 23)
halton_samples_10 = gen_halton_sample(10, validation_space)

print(lhs_samples_10)
ax = plot_sample(lhs_samples_10, 'Validation LHS', 'ob', None, x_range, y_range, 
                x_mark, y_mark, x_label, y_label, 
                'Sampling Using Latin Hypercube Sampling. Sample size: 10')

ax = plot_sample(test, 'Test', 'or', ax)
ax = plot_sample(valid, 'Validation', 'oc', ax)
# ax = plot_sample(halton_samples_10, 'Validation Halton', 'oc', ax)
_ = ax.grid()
_ = ax.legend()
plt.show()
```

```python
# Find the unique sample list
sample_uniq = set()
for x,y in sobol_samples_32:
    sample_uniq.add((x, y))

for x,y in halton_samples_32:
    sample_uniq.add((x, y))
    
for x,y in lhs_samples_32:
    sample_uniq.add((x, y))

for x,y in lhs_samples_24:
    sample_uniq.add((x, y))
    
for x,y in lhs_samples_16:
    sample_uniq.add((x, y))

print(f'Total number of unique samples: {len(sample_uniq)}')
```

### Ensure test and validation data points are not present in training data points

```python
# Check if test data points are present in unique sampels:
for x, y in test:
    if (x, y) in sample_uniq:
        print(f'{x},{y} duplicate test')

for x, y in valid:
    if (x, y) in sample_uniq:
        print(f'{x},{y} duplicate validation')
```

### List All configurations 

```python
num_cycles = []
sizes = []

test_num_cycles = []
test_sizes = []

validation_num_cycles = []
validation_sizes = []

for x, y in sample_uniq:
    num_cycles.append(x)
    sizes.append(y)

for x, y in test:
    test_num_cycles.append(x)
    test_sizes.append(y)
    
for x, y in valid:
    validation_num_cycles.append(x)
    validation_sizes.append(y)

print(f'Training Number of cycles\n{num_cycles}')
print(f'Training sizes\n{sizes}')
print('\n')
print(f'Testing Number of cycles\n{test_num_cycles}')
print(f'Testing sizes\n{test_sizes}')
print('\n')
print(f'Training Number of cycles\n{validation_num_cycles}')
print(f'Training sizes\n{validation_sizes}')
```

# Sampling of Backend Configurations

## Sampling configuration for Axiline design

### Sample space

```python
space = [(0.4, 2.2), (0.4, 0.9)]
x_range = [0.3, 2.3]
y_range = [0.35, 0.95]
x_mark = [0.2, 0.05]
y_mark = [0.1, 0.02]
x_label = 'Frequency'
y_label = "Utilization"
```

### Sampling using LHS

#### Sample trainig configuration

```python
## LHS Sampling
lhs_samples_30 = gen_lhs_sample(30, space, 0, 1, 42)

print(lhs_samples_30)
ax = plot_sample(lhs_samples_30, 'Training', 'ob', None, x_range, y_range, 
                x_mark, y_mark, x_label, y_label, 
                'Backend Config Sampling Using LHS. Sample size: 24')

# ax = plot_sample(test, 'Test', 'or', ax)
# ax = plot_sample(valid, 'Validation', 'oc', ax)
_ = ax.grid()
_ = ax.legend()
plt.show()
```

```python
cp_list = []
util_list = []

for y, x in lhs_samples_30:
    util_list.append(round(x, 3))
    cp_list.append(int(1000/y))

print(f'Backend configuration\nUtilizations\n{util_list}\nClock Periods\n{cp_list}')
```

#### Sample test and validation configuration

```python
## LHS Sampling
valid_bsamples_10 = gen_lhs_sample(10, space, 0, 1, 3)
test_bsamples_10 = gen_lhs_sample(10, space, 0, 1, 23)

print(valid_bsamples_10)
ax = plot_sample(valid_bsamples_10, 'valid', 'ob', None, x_range, y_range, 
                x_mark, y_mark, x_label, y_label, 
                'Backend Config Sampling Using LHS. Sample size: 24')

ax = plot_sample(test_bsamples_10, 'Test', 'or', ax)
# ax = plot_sample(valid, 'Validation', 'oc', ax)
_ = ax.grid()
_ = ax.legend()
plt.show()
```

```python
test_cp_list = []
test_util_list = []

for y, x in test_bsamples_10:
    test_util_list.append(round(x, 3))
    test_cp_list.append(int(1000/y))

valid_cp_list = []
valid_util_list = []

for y, x in valid_bsamples_10:
    valid_util_list.append(round(x, 3))
    valid_cp_list.append(int(1000/y))

print(f'Utilizations\n{test_util_list}\nClock Periods\n{test_cp_list}')
print(f'\nUtilizations\n{valid_util_list}\nClock Periods\n{valid_cp_list}')
```

#### Check if validation and test samples are unique compared to train samples

```python
test_bsample = []
train_bsample = []
valid_bsample = []

unique_list = set()

for i, cp in enumerate(cp_list):
    train_bsample.append([cp, util_list[i]])
    unique_list.add((cp, util_list[i]))

for i, cp in enumerate(test_cp_list):
    _util = test_util_list[i]
    test_bsample.append([cp, _util])
    if (cp, _util) in unique_list:
        print(f'Dplicate test: {cp},{_util}')
    else:
        unique_list.add((cp, _util))

for i, cp in enumerate(valid_cp_list):
    _util = valid_util_list[i]
    valid_bsample.append([cp, _util])
    if (cp, _util) in unique_list:
        print(f'Dplicate valid: {cp},{_util}')
    else:
        unique_list.add((cp, _util))    
```

```python
print(f'Training Backend samples:\n{train_bsample}')
print(f'\nTest Backend samples:\n{test_bsample}')
print(f'\nValid Backend samples:\n{valid_bsample}')
```
